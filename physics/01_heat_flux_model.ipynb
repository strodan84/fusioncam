## Motivation

Plasma-facing components in tokamaks routinely experience heat loads ranging from
approximately 0.5–20 MW/m² depending on plasma conditions.

A thermal imaging diagnostic must estimate

- transient surface temperature
- radiated power
- dynamic range
- frame rate requirements

before hardware can be designed.

This notebook develops a simplified lumped thermal model to estimate these
quantities.

## Heat Pulse

```python
data = hfm.demo_heat_pulse()
```

```python
t = data["t"]
q_in = data["q_in"]
T = data["T"]
q_rad = data["q_rad"]
q_cond = data["q_cond"]
```

```python
plt.figure(figsize=(9,5))

plt.plot(t, q_in/1e6, label="Plasma Heat Flux")
plt.plot(t, q_rad/1e6, label="Radiative Loss")
plt.plot(t, q_cond/1e6, label="Conduction to Bulk")
plt.xlabel("Time (s)")
plt.ylabel("Heat Flux (MW/m²)")
plt.title("Heat Balance During Plasma Pulse")

plt.grid(True)
plt.legend()

plt.show()
```
